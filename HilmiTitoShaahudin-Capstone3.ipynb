Dalam project ini, akan dilakukan rekayasa pada fitur dengan mempertimbangkan kombinasi antara hasil dari metode IQR dan domain knowledge untuk membuat dataset lebih bersih dari outlier. Nantinya, model yang dihasilkan hanya terbatas dapat memprediksi listing baru yang nilai-nilainya pada interval data berdasarkan dataset yang sudah bersih. Diluar interval tersebut, hasil prediksinya akan menjadi bias.

```python
# Cek Outlier
def Cek_Outliers(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    limit = IQR*1.5
    print(f'IQR: {IQR}')


    Upper_fence = Q3 + limit
    Lower_fence = Q1 - limit
    print(f'Limit Bawah: {Lower_fence}')
    print(f'Limit Atas: {Upper_fence}')
    df_outliers = df[(df < Lower_fence) | (df > Upper_fence)] 
    print(f"Jumlah outlier sementara adalah {len(df_outliers)}")
    print(df_outliers)
```

```python
# Outlier pada fitur total_rooms
Cek_Outliers(df_model['total_rooms'])
```

**Benchmark Model**

Akan menggunakan 4 algoritma (Regresi Linear, Ridge, Decision Tree, XGBoost) sebagai benchmarking, dimana nanti dipilih 2 algoritma terbaik untuk dilakukan benchmarking selanjutnya. 

Ringkasan Singkat 4 Model tersebut:

- Regresi Linear: teknik statistik sederhana dan banyak digunakan yang memodelkan hubungan antara variabel dependen (target) dan satu atau lebih variabel independen (fitur) dengan menyesuaikan persamaan linier dengan data yang diamati.

- Ridge: jenis regresi linier yang mencakup istilah regularisasi L2. Regresi ini membantu mencegah overfitting dengan mengecilkan koefisien ke arah nol tetapi tidak tepat nol, yang berguna saat menangani multikolinearitas.

- Decision Tree: algoritma pembelajaran terbimbing nonparametrik yang digunakan untuk klasifikasi dan regresi. Algoritma ini membagi data menjadi beberapa bagian berdasarkan nilai fitur input, sehingga menghasilkan model keputusan seperti pohon.

- XGBoost: XGBoost (Extreme Gradient Boosting) adalah implementasi lanjutan dari algoritma gradient boosting. Algoritma ini membangun sekumpulan pohon keputusan, di mana setiap pohon mengoreksi kesalahan dari pohon sebelumnya.

```python
# Define algoritma yang digunakan
lr = LinearRegression()
rg = Ridge()
dt = DecisionTreeRegressor(random_state=1)
xgb = XGBRegressor(random_state=1)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_rg = TransformedTargetRegressor(rg, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)

# Scaling data
scaler = StandardScaler()

# Kandidat algoritma
models = [log_lr, log_rg, log_dt, log_xgb]

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

# Mencari algoritma terbaik berdasarkan RMSE, MAE, MAPE
for i in models:
    
    crossval = KFold(n_splits=5, shuffle=True, random_state=1)

    estimator = Pipeline([
        ('scaler', scaler),
        ('model', i)
    ])

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
        )

    print(model_cv_rmse, i)

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error', 
        error_score='raise'
        )

    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error', 
        error_score='raise'
        )

    print(model_cv_mape, i)

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())
```

```python
# Evaluasi hasil dari 4 kandidat algoritma yang digunakan
pd.DataFrame({
    'Model': ['Regresi Linear', 'Ridge', 'DecisionTree', 'XGBoost'],
    'Mean_RMSE': nilai_mean_rmse,
    'Std_RMSE': nilai_std_rmse,
    'Mean_MAE': nilai_mean_mae,
    'Std_MAE': nilai_std_mae,
    'Mean_MAPE': nilai_mean_mape,
    'Std_MAPE': nilai_std_mape
})
```

Perbedaan nilai RMSE dan MAE hal yang wajar karena dari formula pun RMSE, nilai residuals atau error dikuadratkan terlebih dahulu, sedangkan MAE hanya perlu di-absolutkan. Hal ini memberikan pinalty lebih besar pada RMSE dibanding MAE.

Selanjutnya, 
- XGBoost adalah model terbaik di antara keempat model yang dievaluasi berdasarkan metrik Mean_RMSE, Mean_MAE, dan Mean_MAPE. Meskipun memiliki variasi RMSE yang sedikit lebih tinggi, kinerja keseluruhannya lebih baik. Sedangkan, Regresi Linear dan Ridge memiliki kinerja yang hampir identik dan merupakan pilihan yang layak, tetapi regresi linear sedikit lebih baik. Decision Tree lebih baik daripada regresi linear dan Ridge, tetapi buruk pada metrik RMSE.

**Tuning**

Tahap ini akan dilakukan tuning dari hyperparameter pada model XGboost untuk meningkatkan performa model. Referensi macam-macam hyperparameter ada di https://xgboost.readthedocs.io/en/latest/parameter.html#general-parameters.

```python
# Jumlah pohon
n_estimators = [100, 250, 500, 750, 1000]

# Kedalaman pohon
max_depth = list(np.arange(3, 22, 3))

# Jumlah Bobot
min_child_weight = [1.5, 10, 15, 20]

# Learning rate
learning_rate = [0.01, 0.03, 0.05, 0.1, 0.3, 0.5, 1]

# Jumlah baris tiap pohon (% dari total baris train set)
subsample = [0.5, 0.6, 0.7, 0.8, 0.9, 1]

# Jumlah feature yang digunakan untuk tiap pohon (% dari total kolom train set)
colsample_bytree = [0.5, 0.6, 0.7, 0.8, 0.9, 1]

# Gamma (min_impurity_decrease)
gamma = [0, 0.25, 0.5, 0.75, 1] # Semakin besar nilainya, semakin konservatif/simpel modelnya

# Alpha (regularization)
reg_alpha = [0, 0.25, 0.5, 0.75, 1] # Semakin besar nilainya, semakin konservatif/simpel modelnya

# Lambda (regularization)
reg_lambda = [0, 0.25, 0.5, 0.75, 1] # Semakin besar nilainya, semakin konservatif/simpel modelnya

# Hyperparam space XGboost
hyperparam_space_xgb = {
    'model__n_estimators': n_estimators,
    'model__max_depth': max_depth, 
    'model__min_child_weight': min_child_weight,
    'model__learning_rate': learning_rate,
    'model__subsample': subsample,
    'model__colsample_bytree': colsample_bytree,
    'model__gamma': gamma,
    'model__reg_alpha': reg_alpha,
    'model__reg_lambda': reg_lambda
}
```

```python
# Benchmark model dengan hyperparameter tuning
xgb = XGBRegressor(random_state=1, verbosity=0)

# Membuat algorithm chains
estimator_xgb = Pipeline([
        ('scaler', scaler),
        ('model', xgb)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=1)

# Hyperparameter tuning
xgb_tuned = RandomizedSearchCV(
    estimator_xgb, 
    param_distributions = hyperparam_space_xgb,
    cv = crossval, 
    scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'], 
    n_jobs = -1,
    refit = 'neg_root_mean_squared_error', # Hanya bisa memilih salah stau metric untuk optimisasi
    random_state = 1  
)
```

1. Menggunakan Multivariate Testing (MVT)
- Tujuan: Menguji beberapa variabel sekaligus untuk melihat kombinasi mana yang berkinerja terbaik.
- Pros: Memberikan insight lebih komprehensif nantinya karena mengetahui kombinasi fitur terbaik.

- Pros: Hasil tuning memberikan hasil yang signifikan dibanding machine learning.
- Cons: Data harus lebih kompleks (lebih banyak jumlah dan fiturnya). Jika data masih seperti kasus ini, tidak akan memberikan hasil yang signifikan.
